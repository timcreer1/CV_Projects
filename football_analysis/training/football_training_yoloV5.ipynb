```python
import shutil
import os

# Define the source and destination paths
source_path = '/content/football-players-detection-1'
destination_path = '/content/drive/MyDrive/Colab Notebooks/2024 Data Science Projects/Football Analysis/training/football-players-detection-1'

# Move the file
shutil.move(source_path, destination_path)

#Update the data.yaml paths
data_yaml_path = os.path.join(destination_path, 'data.yaml')
with open(data_yaml_path, 'r') as file:
    data_yaml = file.read()

# Replace the relative paths with absolute paths
data_yaml = data_yaml.replace(
    'football-players-detection-1/train/images',
    f'{destination_path}/train/images')
data_yaml = data_yaml.replace(
    'football-players-detection-1/valid/images',
    f'{destination_path}/valid/images')
data_yaml = data_yaml.replace(
    '../test/images',
    f'{destination_path}/test/images')

with open(data_yaml_path, 'w') as file:
    file.write(data_yaml)
```

# Training

```python
from ultralytics import YOLO

#Set paths and model parameters
dataset_path_yaml = '/content/drive/MyDrive/Colab Notebooks/2024 Data Science Projects/Football Analysis/training/football-players-detection-1/data.yaml'
model_name = 'yolov8m.pt'  # Use 'yolov8x.pt' for the largest model
epochs = 100
img_size = 640
model = YOLO('yolov8m.pt')
# Load and train the YOLOv8 model
YOLO(model_name).train(
    data=dataset_path_yaml,
    epochs=epochs,
    imgsz=img_size,
    project='/content/drive/MyDrive/Colab Notebooks/2024 Data Science Projects/Football Analysis/training/results',
    device='cuda'  # Use GPU if available
)
```
